````python
import os
import io
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import google.generativeai as genai
from typing import Optional, Dict, List, Tuple, Any
import pdfplumber  # PDF processing
from PIL import Image
import pytesseract  # OCR capability
import base64
from IPython.display import display, Markdown, HTML
import json
from google.colab import files
import ipywidgets as widgets

class PitchDeckAnalyzer:
    """
    A class that analyzes startup pitch decks using Google's Gemini API.
    Extracts text, evaluates key sections, and provides scoring and feedback.
    """

    def __init__(self, api_key: Optional[str] = None, model_name: str = 'gemini-1.5-pro'):
        """
        Initialize the analyzer with API credentials and model configuration

        Args:
            api_key: Gemini API key (optional if already configured in environment)
            model_name: The specific Gemini model to use
        """
        # Configure the Gemini API
        if api_key:
            genai.configure(api_key=api_key)
        else:
            # This assumes API key is configured in environment variables
            try:
                genai.configure()
            except Exception as e:
                raise ValueError(f"No API key provided and no environment configuration found: {e}")

        # Initialize the Gemini model
        try:
            self.model = genai.GenerativeModel(model_name)
            self.model_name = model_name
        except Exception as e:
            raise ValueError(f"Failed to initialize Gemini model '{model_name}': {e}")

        # Define the sections we want to analyze in a pitch deck
        self.sections = {
            "problem": {
                "weight": 0.15,
                "description": "Problem the startup is solving"
            },
            "solution": {
                "weight": 0.20,
                "description": "The startup's solution to the problem"
            },
            "market": {
                "weight": 0.15,
                "description": "Market size, trends, and opportunity"
            },
            "business_model": {
                "weight": 0.15,
                "description": "How the startup plans to make money"
            },
            "financials": {
                "weight": 0.15,
                "description": "Financial projections and metrics"
            },
            "team": {
                "weight": 0.10,
                "description": "Founding team and key personnel"
            },
            "overall": {
                "weight": 0.10,
                "description": "Overall coherence and quality of the pitch"
            }
        }

    def extract_text_from_pdf(self, pdf_path: str) -> Dict[int, str]:
        """
        Extract text from a PDF file using pdfplumber

        Args:
            pdf_path: Path to the PDF file

        Returns:
            Dictionary mapping page numbers to extracted text
        """
        text_by_page = {}

        try:
            # Open the PDF with pdfplumber
            with pdfplumber.open(pdf_path) as pdf:
                # Extract text from each page
                for page_num, page in enumerate(pdf.pages):
                    text = page.extract_text() or ""

                    # If text extraction yields minimal results, try OCR
                    if len(text.strip()) < 50:  # Arbitrary threshold
                        # Convert page to image
                        img = page.to_image(resolution=300)
                        # Save image to a temporary bytes buffer
                        img_bytes = io.BytesIO()
                        img.save(img_bytes, format='PNG')
                        img_bytes.seek(0)
                        # Apply OCR
                        pil_img = Image.open(img_bytes)
                        text = pytesseract.image_to_string(pil_img)

                    text_by_page[page_num] = text

            return text_by_page

        except Exception as e:
            raise Exception(f"Error extracting text from PDF: {e}")

    def preprocess_text(self, text_by_page: Dict[int, str]) -> str:
        """
        Preprocess the extracted text by cleaning and normalizing

        Args:
            text_by_page: Dictionary of page numbers to text

        Returns:
            Preprocessed text as a single string
        """
        combined_text = ""

        for page_num in sorted(text_by_page.keys()):
            page_text = text_by_page[page_num]

            # Remove excessive whitespace
            page_text = re.sub(r'\s+', ' ', page_text)

            # Remove common header/footer patterns
            page_text = re.sub(r'Page \d+ of \d+', '', page_text)

            # Add a page separator
            combined_text += f"\n--- PAGE {page_num + 1} ---\n{page_text}\n"

        return combined_text

    def identify_sections(self, text: str) -> Dict[str, str]:
        """
        Use Gemini to identify key sections in the pitch deck text

        Args:
            text: Preprocessed text from the pitch deck

        Returns:
            Dictionary mapping section names to extracted content
        """
        prompt = f"""
        You are an expert pitch deck analyzer. I'm going to provide you with the text extracted from a startup pitch deck.
        Your task is to identify and extract the following key sections:

        1. Problem: What problem is the startup solving?
        2. Solution: What solution is the startup offering?
        3. Market: Information about market size, trends, and opportunity
        4. Business Model: How the startup plans to make money
        5. Financials: Financial projections and metrics
        6. Team: Information about the founding team and key personnel

        For each section, extract the relevant text. If a section appears to be missing, indicate that with 'Not explicitly covered in the pitch deck.'

        Format your response as JSON with keys: 'problem', 'solution', 'market', 'business_model', 'financials', 'team'.

        Here is the pitch deck text:

        {text}
        """

        try:
            response = self.model.generate_content(prompt)
            response_text = response.text

            # Extract JSON from the response
            json_match = re.search(r'```json\s*(.*?)\s*```', response_text, re.DOTALL)
            if json_match:
                json_str = json_match.group(1)
            else:
                # If not in code block, try to find JSON directly
                json_str = response_text

            # Clean up and parse JSON
            sections = json.loads(json_str)

            # Ensure all required sections exist
            for section in ['problem', 'solution', 'market', 'business_model', 'financials', 'team']:
                if section not in sections:
                    sections[section] = "Not explicitly covered in the pitch deck."

            return sections

        except Exception as e:
            raise Exception(f"Error identifying sections: {e}")

    def evaluate_section(self, section_name: str, section_content: str) -> Dict[str, Any]:
        """
        Evaluate a specific section of the pitch deck using Gemini

        Args:
            section_name: Name of the section being evaluated
            section_content: Text content of the section

        Returns:
            Dictionary with evaluation results including score and feedback
        """
        section_descriptions = {
            "problem": "clearly defines a significant pain point that resonates with the target audience",
            "solution": "presents a unique, innovative solution with clear value proposition and competitive advantages",
            "market": "demonstrates deep understanding of the market size, trends, and growth potential with credible data",
            "business_model": "presents a clear, sustainable revenue model with realistic customer acquisition strategy",
            "financials": "provides realistic financial projections with clear assumptions and understanding of unit economics",
            "team": "showcases a strong, complementary team with relevant experience and expertise for the venture",
            "overall": "presents a coherent, compelling narrative with professional design and clarity of communication"
        }

        prompt = f"""
        You are an expert venture capitalist evaluating a startup pitch deck. Analyze this {section_name} section and provide:

        1. A score from 0-100 based on how well the {section_name} section {section_descriptions.get(section_name, "is presented")}
        2. Key strengths (2-3 points)
        3. Areas for improvement (2-3 points)
        4. Specific, actionable recommendations

        Section content:
        {section_content}

        Format your response as a JSON object with the following structure:
        {{
          "score": <0-100 integer>,
          "strengths": ["strength 1", "strength 2", ...],
          "weaknesses": ["weakness 1", "weakness 2", ...],
          "recommendations": ["recommendation 1", "recommendation 2", ...]
        }}

        Do not include any text outside the JSON object.
        """

        try:
            response = self.model.generate_content(prompt)
            response_text = response.text

            # Extract JSON from the response
            json_match = re.search(r'```json\s*(.*?)\s*```', response_text, re.DOTALL)
            if json_match:
                json_str = json_match.group(1)
            else:
                # If not in code block, try to find JSON directly
                json_str = response_text

            # Clean up and parse JSON
            evaluation = json.loads(json_str)

            return evaluation

        except Exception as e:
            # If evaluation fails, return a default with error message
            return {
                "score": 0,
                "strengths": ["Could not evaluate"],
                "weaknesses": [f"Error during evaluation: {str(e)}"],
                "recommendations": ["Ensure the section is clearly defined and contains relevant information"]
            }

    def evaluate_overall_pitch(self, sections: Dict[str, str], section_evaluations: Dict[str, Dict]) -> Dict[str, Any]:
        """
        Evaluate the overall pitch deck based on extracted sections and individual evaluations

        Args:
            sections: Dictionary of section contents
            section_evaluations: Dictionary of section evaluation results

        Returns:
            Overall evaluation including total score and summary feedback
        """
        # Construct context for the overall evaluation
        context = "Pitch Deck Sections Summary:\n\n"
        for section_name, content in sections.items():
            section_score = section_evaluations.get(section_name, {}).get("score", 0)
            context += f"{section_name.upper()} (Score: {section_score}/100):\n{content[:200]}...\n\n"

        prompt = f"""
        You are an expert venture capitalist who has reviewed the key sections of a startup pitch deck.
        Based on the following summary of the pitch deck sections and their individual scores, provide:

        1. An overall assessment of the pitch deck's effectiveness (0-100)
        2. The key strengths of the overall pitch (3 points)
        3. The key weaknesses of the overall pitch (3 points)
        4. Specific recommendations for improving the pitch deck as a whole
        5. A brief assessment of the startup's fundraising readiness

        {context}

        Format your response as a JSON object with the following structure:
        {{
          "score": <0-100 integer>,
          "strengths": ["strength 1", "strength 2", "strength 3"],
          "weaknesses": ["weakness 1", "weakness 2", "weakness 3"],
          "recommendations": ["recommendation 1", "recommendation 2", "recommendation 3"],
          "fundraising_readiness": "<brief assessment>"
        }}

        Do not include any text outside the JSON object.
        """

        try:
            response = self.model.generate_content(prompt)
            response_text = response.text

            # Extract JSON from the response
            json_match = re.search(r'```json\s*(.*?)\s*```', response_text, re.DOTALL)
            if json_match:
                json_str = json_match.group(1)
            else:
                # If not in code block, try to find JSON directly
                json_str = response_text

            # Clean up and parse JSON
            overall_evaluation = json.loads(json_str)

            return overall_evaluation

        except Exception as e:
            # If evaluation fails, return a default with error message
            return {
                "score": 0,
                "strengths": ["Could not evaluate overall pitch"],
                "weaknesses": [f"Error during evaluation: {str(e)}"],
                "recommendations": ["Review the entire pitch deck for clarity and completeness"],
                "fundraising_readiness": "Unable to assess due to evaluation error"
            }

    def calculate_weighted_score(self, section_evaluations: Dict[str, Dict], overall_evaluation: Dict[str, Any]) -> float:
        """
        Calculate the final weighted score based on section scores and weights

        Args:
            section_evaluations: Dictionary of section evaluation results
            overall_evaluation: Overall pitch evaluation

        Returns:
            Final weighted score (0-100)
        """
        total_score = 0

        # Add weighted section scores
        for section_name, section_info in self.sections.items():
            if section_name == "overall":
                section_score = overall_evaluation.get("score", 0)
            else:
                section_score = section_evaluations.get(section_name, {}).get("score", 0)

            weight = section_info["weight"]
            total_score += section_score * weight

        return round(total_score, 1)

    def generate_visualizations(self, section_evaluations: Dict[str, Dict], overall_score: float) -> Dict[str, Any]:
        """
        Generate visualizations for the pitch deck analysis

        Args:
            section_evaluations: Dictionary of section evaluation results
            overall_score: Final weighted score

        Returns:
            Dictionary of visualization data
        """
        # Prepare data for visualizations
        sections = []
        scores = []

        for section_name, evaluation in section_evaluations.items():
            sections.append(section_name.capitalize())
            scores.append(evaluation.get("score", 0))

        # Create radar chart data
        radar_data = {
            'sections': sections,
            'scores': scores
        }

        # Create bar chart data
        bar_data = {
            'sections': sections + ['Overall'],
            'scores': scores + [overall_score]
        }

        return {
            'radar_data': radar_data,
            'bar_data': bar_data
        }

    def plot_visualizations(self, visualizations: Dict[str, Any]):
        """
        Plot the visualizations for the pitch deck analysis

        Args:
            visualizations: Dictionary of visualization data
        """
        # Set style
        plt.style.use('ggplot')

        # Create figure with two subplots - the first one with polar projection for radar chart
        fig = plt.figure(figsize=(16, 7))
        ax1 = fig.add_subplot(121, projection='polar')  # This creates a polar axes for the radar chart
        ax2 = fig.add_subplot(122)  # Regular axes for bar chart

        # Plot radar chart
        radar_data = visualizations['radar_data']

        # Number of variables
        N = len(radar_data['sections'])

        # What will be the angle of each axis in the plot
        angles = [n / float(N) * 2 * np.pi for n in range(N)]
        angles += angles[:1]  # Close the loop

        # Add scores
        scores = radar_data['scores']
        scores += scores[:1]  # Close the loop

        # Draw the axes
        ax1.set_theta_offset(np.pi / 2)  # This will work now with polar projection
        ax1.set_theta_direction(-1)

        # Draw one axis per variable and add labels
        ax1.set_xticks(angles[:-1])
        ax1.set_xticklabels(radar_data['sections'], color='grey', size=10)

        # Draw the chart
        ax1.plot(angles, scores, linewidth=2, linestyle='solid')
        ax1.fill(angles, scores, alpha=0.4)

        # Add a title
        ax1.set_title("Pitch Deck Section Scores", size=15, color='grey', y=1.1)

        # Set y-axis limits
        ax1.set_ylim(0, 100)

        # Plot bar chart
        bar_data = visualizations['bar_data']
        sns.barplot(x='sections', y='scores', data=pd.DataFrame(bar_data), ax=ax2)
        ax2.set_xlabel('Sections')
        ax2.set_ylabel('Scores')
        ax2.set_title('Section and Overall Scores')
        ax2.set_ylim(0, 100)
        ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')

        plt.tight_layout()
        plt.show()

    def analyze_pitch_deck(self, pdf_path: str) -> Dict[str, Any]:
        """
        Analyze a pitch deck and provide scoring and feedback

        Args:
            pdf_path: Path to the pitch deck PDF file

        Returns:
            Dictionary containing analysis results
        """
        # Extract and preprocess text
        print("ðŸ“„ Extracting text from PDF...")
        text_by_page = self.extract_text_from_pdf(pdf_path)
        processed_text = self.preprocess_text(text_by_page)

        # Identify sections
        print("ðŸ” Identifying key sections...")
        sections = self.identify_sections(processed_text)

        # Evaluate each section
        print("âš–ï¸ Evaluating sections...")
        section_evaluations = {}
        for section_name, section_content in sections.items():
            print(f"  - Evaluating {section_name} section...")
            section_evaluations[section_name] = self.evaluate_section(section_name, section_content)

        # Evaluate overall pitch
        print("ðŸ† Evaluating overall pitch...")
        overall_evaluation = self.evaluate_overall_pitch(sections, section_evaluations)

        # Calculate final score
        final_score = self.calculate_weighted_score(section_evaluations, overall_evaluation)

        # Generate visualizations
        visualizations = self.generate_visualizations(section_evaluations, final_score)

        # Prepare results
        results = {
            'pitch_deck_path': pdf_path,
            'sections': sections,
            'section_evaluations': section_evaluations,
            'overall_evaluation': overall_evaluation,
            'final_score': final_score,
            'visualizations': visualizations
        }

        return results

    def display_analysis_results(self, results: Dict[str, Any]):
        """
        Display the analysis results in a readable format

        Args:
            results: Dictionary containing analysis results
        """
        # Plot visualizations
        self.plot_visualizations(results['visualizations'])

        # Print results as Markdown
        markdown_output = f"""
        # Pitch Deck Analysis Results

        ## Overall Score: {results['final_score']}/100

        ### Fundraising Readiness
        {results['overall_evaluation'].get('fundraising_readiness', 'Not assessed')}

        ## Overall Evaluation

        ### Strengths
        {self._format_list(results['overall_evaluation'].get('strengths', []))}

        ### Areas for Improvement
        {self._format_list(results['overall_evaluation'].get('weaknesses', []))}

        ### Recommendations
        {self._format_list(results['overall_evaluation'].get('recommendations', []))}

        ## Section Evaluations
        """

        for section_name, evaluation in results['section_evaluations'].items():
            markdown_output += f"""
        ### {section_name.capitalize()} (Score: {evaluation.get('score', 0)}/100)

        #### Strengths
        {self._format_list(evaluation.get('strengths', []))}

        #### Areas for Improvement
        {self._format_list(evaluation.get('weaknesses', []))}

        #### Recommendations
        {self._format_list(evaluation.get('recommendations', []))}
            """

        display(Markdown(markdown_output))

    def _format_list(self, items: List[str]) -> str:
        """Helper method to format list items for Markdown display"""
        if not items:
            return "- None identified"

        return "\n".join([f"- {item}" for item in items])

    def save_analysis_report(self, results: Dict[str, Any], output_path: str):
        """
        Save the analysis results to a file (PDF or Markdown)

        Args:
            results: Dictionary containing analysis results
            output_path: Path to save the report
        """
        # Create Markdown content
        markdown_content = f"""
# Pitch Deck Analysis Report

## Executive Summary

**Overall Score:** {results['final_score']}/100

**Fundraising Readiness:** {results['overall_evaluation'].get('fundraising_readiness', 'Not assessed')}

## Overall Evaluation

### Strengths
{self._format_list(results['overall_evaluation'].get('strengths', []))}

### Areas for Improvement
{self._format_list(results['overall_evaluation'].get('weaknesses', []))}

### Recommendations
{self._format_list(results['overall_evaluation'].get('recommendations', []))}

## Section Evaluations
"""

        for section_name, evaluation in results['section_evaluations'].items():
            markdown_content += f"""
### {section_name.capitalize()} (Score: {evaluation.get('score', 0)}/100)

#### Strengths
{self._format_list(evaluation.get('strengths', []))}

#### Areas for Improvement
{self._format_list(evaluation.get('weaknesses', []))}

#### Recommendations
{self._format_list(evaluation.get('recommendations', []))}
"""

        # Save to file
        with open(output_path, 'w') as f:
            f.write(markdown_content)

        print(f"Report saved to {output_path}")

# Example usage in a notebook
if __name__ == "__main__":
    # This will run if the file is executed as a script
    print("Pitch Deck Analyzer initialized. Use the 'analyze_pitch_deck' method to analyze a pitch deck.")

    # Step 1: Set up your API key
    # Replace with your actual Gemini API key
    GEMINI_API_KEY = ""  # Or use os.environ.get("GEMINI_API_KEY")

    # Step 2: Initialize the analyzer
    analyzer = PitchDeckAnalyzer(api_key=GEMINI_API_KEY)

    # Step 3: Analyze a single pitch deck
    print("Analyzing Sample Pitch Deck 1...")
    sample_pitch_deck_path = "/content/uber-pitch-deck.pdf"  # Replace with your actual file path
    results = analyzer.analyze_pitch_deck(sample_pitch_deck_path)

    # Step 4: Display the results
    analyzer.display_analysis_results(results)

    # Step 5: Save the report
    analyzer.save_analysis_report(results, "company1_pitch_analysis.md")

    # Step 6: Create an upload widget for user pitch decks
    def upload_and_analyze():
        """Function to handle pitch deck upload and analysis"""
        print("Please upload your pitch deck PDF...")
        uploaded = files.upload()

        if uploaded:
            filename = list(uploaded.keys())[0]
            print(f"Analyzing uploaded pitch deck: {filename}")
            results = analyzer.analyze_pitch_deck(filename)
            analyzer.display_analysis_results(results)
            analyzer.save_analysis_report(results, f"{os.path.splitext(filename)[0]}_analysis.md")
            print("Analysis complete! Download the report using the link above.")

    # Create an upload button
    upload_button = widgets.Button(
        description='Upload & Analyze Your Pitch Deck',
        button_style='primary',
        icon='upload'
    )
    upload_button.on_click(lambda b: upload_and_analyze())
    display(upload_button)
````
